## **Exercise 2a:**

```python
from qiskit import QuantumCircuit
from qiskit import QuantumRegister, QuantumCircuit
import math
pi=math.pi
c = QuantumRegister(1, 'control')
t = QuantumRegister(2, 'target')
cu = QuantumCircuit(c, t, name="Controlled 13^x mod 35")

# WRITE YOUR CODE BETWEEN THESE LINES - START
#from math import pi
# WRITE YOUR CODE BETWEEN THESE LINES - START

# Controlled Controlled NOT
#Hadamard
cu.u(pi/2, 0, 0, 2)
# X gate
cu.u(pi , 0 , pi, 2)
cu.cx(1,2)
#cu.tdg(2)
cu.u(0, 0, -pi/4, 2)
cu.cx(0,2)
cu.u(0, 0, pi/4, 2)
cu.cx(1,2)
cu.u(0, 0, -pi/4, 2)
cu.cx(0,2)
cu.u(0, 0, pi/4, 1)
cu.u(0, 0, pi/4, 2)
#Hadamard
cu.u(pi/2, 0, 0, 2)
# X gate
cu.u(pi , 0 , pi, 2)
cu.cx(0,1)
cu.u(0, 0, pi/4, 0)
cu.u(0, 0, -pi/4, 1)
cu.cx(0,1)
# Control NOT
cu.cx(0,1)
# WRITE YOUR CODE BETWEEN THESE LINES - END
cu.draw('mpl')
```
